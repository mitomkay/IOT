```python
# Tạo danh sách các mẫu huấn luyện
X_train = []
y_train = []

# Tạo danh sách các mẫu kiểm tra
X_test = []
y_test = []

# Tổ chức dữ liệu thành các nhóm 24 phần tử liên tiếp
for i in range(0, len(X), 24):
    X_group = X.iloc[i:i+24]
    y_group = y.iloc[i//24]  # Chỉ lấy nhãn đầu tiên trong mỗi nhóm

    # Chia dữ liệu thành tập huấn luyện và tập kiểm tra
    if i % 4 == 0:
        X_test.append(X_group)
        y_test.append(y_group)
    else:
        X_train.append(X_group)
        y_train.append(y_group)

# Gộp danh sách thành DataFrame
X_train = pd.concat(X_train, ignore_index=True)
y_train = pd.Series(y_train)
X_test = pd.concat(X_test, ignore_index=True)
y_test = pd.Series(y_test)

# Kiểm tra xem danh sách không rỗng trước khi huấn luyện mô hình
if X_train.empty or y_train.empty:
    print("Không có dữ liệu huấn luyện")
    exit()

# Khởi tạo mô hình RandomForestClassifier
model = RandomForestClassifier()

# Huấn luyện mô hình trên tập huấn luyện
model.fit(X_train, y_train)

# Dự đoán nhãn cho tập kiểm tra
if not X_test.empty:
    y_pred = model.predict(X_test)

    # Đánh giá độ chính xác
    accuracy = accuracy_score(y_test, y_pred)
    print("Độ chính xác của mô hình:", accuracy)
else:
    print("Không có dữ liệu kiểm tra")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Chia dữ liệu thành tập huấn luyện và tập kiểm tra (theo tỷ lệ 60:40)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.60, shuffle=True, random_state=0)

# Tạo danh sách các mẫu huấn luyện
X_train_grouped = []
y_train_grouped = []

# Tạo danh sách các mẫu kiểm tra
X_test_grouped = []
y_test_grouped = []

# Tổ chức dữ liệu thành các nhóm 24 phần tử liên tiếp
for i in range(0, len(X_train), 24):
    X_group = X_train.iloc[i:i+24]
    y_group = y_train.iloc[i//24]  # Chỉ lấy nhãn đầu tiên trong mỗi nhóm

    # Thực hiện các xử lý dữ liệu khác tại đây (nếu cần)

    X_train_grouped.append(X_group)
    y_train_grouped.append(y_group)

# Tổ chức dữ liệu kiểm tra thành các nhóm 24 phần tử liên tiếp
for i in range(0, len(X_test), 24):
    X_group = X_test.iloc[i:i+24]
    y_group = y_test.iloc[i//24]  # Chỉ lấy nhãn đầu tiên trong mỗi nhóm

    # Thực hiện các xử lý dữ liệu khác tại đây (nếu cần)

    X_test_grouped.append(X_group)
    y_test_grouped.append(y_group)

# Khởi tạo StandardScaler
scaler = StandardScaler()

# Huấn luyện scaler trên tập huấn luyện đã được nhóm
scaler.fit(pd.concat(X_train_grouped))

# Chuẩn hóa dữ liệu huấn luyện và kiểm tra theo scaler đã huấn luyện
X_train_scaled = [pd.DataFrame(scaler.transform(X), columns=X.columns) for X in X_train_grouped]
X_test_scaled = [pd.DataFrame(scaler.transform(X), columns=X.columns) for X in X_test_grouped]

# Gán lại nhãn cho các nhóm dữ liệu
y_train_grouped = pd.Series(y_train_grouped)
y_test_grouped = pd.Series(y_test_grouped)

# Gộp các nhóm dữ liệu thành DataFrame
X_train = pd.concat(X_train_scaled, ignore_index=True)
X_test = pd.concat(X_test_scaled, ignore_index=True)
y_train = pd.Series(y_train_grouped)
y_test = pd.Series(y_test_grouped)

X_train, X_test, y_train, y_test
```

```python
import numpy as np

# Tạo dữ liệu mẫu
X = np.random.randn(100)  # 100 phần tử dữ liệu đầu vào
y = np.random.randint(0, 2, size=5)  # 5 nhãn y

# Tạo danh sách các nhóm dữ liệu
X_grouped = []
y_grouped = []

for i in range(0, len(X), 24):
    X_group = X[i:i+24]
    y_group = y[i//24]

    X_grouped.append(X_group)
    y_grouped.append(y_group)

# In kết quả
for X_group, y_group in zip(X_grouped, y_grouped):
    print("X:", X_group)
    print("y:", y_group)
    print("---")
```

```python
from sklearn.linear_model import LinearRegression

models = []  # Danh sách các mô hình tương ứng với từng nhóm dữ liệu

# Huấn luyện mô hình trên từng nhóm dữ liệu
for X_group, y_group in zip(X_grouped, y_grouped):
    X_group = np.array(X_group).reshape(-1, 1)
    y_group = np.array(y_group).reshape(-1, 1)
    
    model = LinearRegression()
    model.fit(X_group, y_group)
    models.append(model)

# Sau khi huấn luyện, bạn có thể sử dụng mô hình để dự đoán các giá trị mới
# Ví dụ: Dự đoán giá trị y cho một mẫu dữ liệu X_new
X_new = np.random.randn(24)  # Một mẫu dữ liệu mới
X_new = X_new.reshape(-1, 1)  # Reshape X_new thành (n_samples, n_features)

for model in models:
    y_pred = model.predict(X_new)
    print("Dự đoán y:", y_pred)
```

```python
import numpy as np
import pandas as pd
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import joblib

X = pd.read_csv('data.csv')
y = pd.read_csv('label.csv')

def preprocessing(df, task):
    if task == 'Regression':
        Y = y['area'].values
    elif task == 'Classification':
        Y = y['area'].apply(lambda x: 1 if x > 0 else 0).values

    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.60, shuffle=True, random_state=0)

    scaler = StandardScaler()
    scaler.fit(X_train)

    X_train = pd.DataFrame(scaler.transform(X_train), columns=X.columns)
    X_test = pd.DataFrame(scaler.transform(X_test), columns=X.columns)

    return X_train, X_test, Y_train, Y_test


X_train, X_test, Y_train, Y_test = preprocessing(X, task='Regression')

nn_regressor_model = MLPRegressor(activation='relu', hidden_layer_sizes=(16, 16), max_iter=100, solver='adam')
nn_regressor_model.fit(X_train, Y_train)

model = joblib.dump(nn_regressor_model, 'forestfiremodel.pkl')
```
